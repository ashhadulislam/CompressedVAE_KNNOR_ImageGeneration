```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import os
import random
```

```python
real_images_location="Data/celeba/img_align_celeba/"
real_image_files=os.listdir(real_images_location)
print(real_image_files[10])
```

```python
model_names=["VanillaVAE","ConditionalVAE","DFCVAE","BetaVAE","MSSIMVAE","WAE_MMD"]
prune_rates=["0% compressed","~15% compressed","~50% compressed"]

all_img_locations=[]
for model_name in model_names:
    loc_1="logs/"+model_name+"/artificial/all_imgs/pruned/"
    prue_rate_folders=os.listdir(loc_1)
    prue_rate_folders.sort()
    row_files=[]
    # choose a real image file
    chosen_real_image=real_images_location+random.sample(real_image_files,1)[0]
    row_files.append(chosen_real_image)
    for prune_fold in prue_rate_folders:    
        loc_2=loc_1+prune_fold+"/"
#         print(prune_fold,loc_2)
        all_pruned_images=os.listdir(loc_2)
#         print(len(all_pruned_images))
        chosen_pruned_image=loc_2+random.sample(all_pruned_images,1)[0]
        row_files.append(chosen_pruned_image)
        # choose an image randomly
    all_img_locations.append(row_files)
```

```python
all_img_locations
```

```python
rows=6
cols=4
fig,axs=plt.subplots(rows,cols,figsize=(15, 15))
plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
for i in range(rows):
    for j in range(cols):  
        im=Image.open(all_img_locations[i][j])        
        if i == 0:
            if j==0:
                axs[i][j].title.set_text('Real Image')
            else:
                axs[i][j].title.set_text(prune_rates[j-1])
        axs[i][j].imshow(im)
fig.tight_layout()
plt.savefig("plt_imgs/diff_prunes.png",bbox_inches='tight')
plt.savefig("plt_imgs/diff_prunes.pdf",bbox_inches='tight')
plt.savefig("plt_imgs/diff_prunes.svg",bbox_inches='tight')
```

### Now for GAN

```python
gan_images_location="GANops/all_imgs/"
gan_image_files=os.listdir(gan_images_location)
```

```python
model_names=["VanillaVAE","ConditionalVAE","DFCVAE","BetaVAE","MSSIMVAE","WAE_MMD"]
prune_rates=["0% compressed","~15% compressed","~50% compressed"]
model_names_pic=["VBAE","CVAE","DFC-VAE","Beta-VAE","MSSIM-VAE","WAE-MMD"]
all_img_locations=[]
for model_name in model_names:
    loc_1="logs/"+model_name+"/artificial/all_imgs/pruned/"
    prue_rate_folders=os.listdir(loc_1)
    prue_rate_folders.sort()
    row_files=[]
    # choose a real image file
    chosen_real_image=real_images_location+random.sample(real_image_files,1)[0]    
    row_files.append(chosen_real_image)
    # choose a gan image file
    chosen_gan_image=gan_images_location+random.sample(gan_image_files,1)[0]    
    row_files.append(chosen_gan_image)
    
    for prune_fold in prue_rate_folders[:1]:    
        loc_2=loc_1+prune_fold+"/"
#         print(prune_fold,loc_2)
        all_pruned_images=os.listdir(loc_2)
#         print(len(all_pruned_images))
        chosen_pruned_image=loc_2+random.sample(all_pruned_images,1)[0]
        row_files.append(chosen_pruned_image)
        # choose an image randomly
    all_img_locations.append(row_files)
```

```python
all_img_locations
```

```python
rows=6
cols=3
fig,axs=plt.subplots(rows,cols,figsize=(15, 15))
plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
for i in range(rows):
    for j in range(cols):  
        im=Image.open(all_img_locations[i][j])        
        if i == 0:
            if j==0:
                axs[i][j].title.set_text('Real Image')            
            elif j==1:
                axs[i][j].title.set_text('DCGAN Image')                            
        
        if j==2:
            axs[i][j].title.set_text(model_names_pic[i])                            
                
                
        axs[i][j].imshow(im)
fig.tight_layout()
plt.savefig("plt_imgs/realVGANvVK.png",bbox_inches='tight')
plt.savefig("plt_imgs/realVGANvVK.pdf",bbox_inches='tight')
plt.savefig("plt_imgs/realVGANvVK.svg",bbox_inches='tight')
```
