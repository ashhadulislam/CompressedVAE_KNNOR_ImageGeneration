```python
# load the required packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image into system memory
image = cv2.imread('logs/VanillaVAE/artificial/all_imgs/pruned/0/img20.jpeg', flags=cv2.IMREAD_COLOR)
image_arr=np.array(image)
```

```python
plt.hist(image_arr.flatten())
```

```python
plt.imshow(image)
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
image_arr_sharp=np.array(image_sharp)
```

```python
plt.imshow(image_sharp)
```

```python
plt.hist(image_arr_sharp.flatten())
```
